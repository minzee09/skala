## 해결하려는 문제

**회의 이후의 반복 작업은 여전히 사람의 몫이며, 이를 대체할 자동화 도구는 보안이 중요한 조직에서는 사용할 수 없다.**

회의가 끝나면 다음 세 가지 작업이 남는다.

① 담당자가 녹음을 다시 들으며 회의록을 정리한다. 회의 한 건당 약 20분에서 30분이 소요된다.

② 회의 중 언급된 회의실과 후속 일정을 별도 시스템에서 수동으로 예약해야 하며, 이 과정에서 누락이 발생한다.

③ 과거 논의를 확인하려면 녹음을 재생하거나 회의록을 일일이 열어봐야 한다.

이를 자동화하는 제품은 이미 여럿 있으나 대부분 음성과 텍스트를 외부 클라우드로 전송해 처리한다. 회의록에는 결론만 남지만 발화록에는 채택되지 않은 대안, 협상 가능 범위, 대외 미공개 일정이 그대로 담긴다. 정리된 문서보다 민감도가 높은 원본이 외부로 나가는 셈이다.

저장소만 내부에 두는 방식으로는 해결되지 않는다. 요약을 하려면 발화록을 모델에 입력해야 하므로, 모델이 외부에 있으면 원문은 이미 외부를 거친 뒤다. 도입 가능 여부를 가르는 것은 데이터의 저장 위치가 아니라 추론이 어디에서 일어나는지다.

또한 자동 생성된 요약은 어느 발언, 어느 회의에서 나온 내용인지 확인할 수 없다. 생성 모델은 다음 토큰의 확률 분포에서 토큰을 선택하므로 발화록에 없는 내용을 만들어낼 가능성이 원리적으로 남으며, 이를 판별할 장치가 없으면 그대로 기록이 된다는 문제점이 존재한다.

---

## 프로젝트 목표

회의 발화록을 입력받아 회의록, 액션아이템, 다음 회의실 예약 초안을 생성하고, 이후 발화록을 근거로 자연어 질문에 답하는 에이전트 시스템을 구현하는 것까지 목표로 한다.

생성과 응답 양쪽 모두에서 모든 문장에 근거 발화 ID를 부착하고, 생성이 끝난 뒤 코드로 그 인용이 실재하는지 대조하여 요약 결과를 원문 발화로 역추적할 수 있게 만드는 것이 목표다.

또한, 추론이 조직 내부에서 완결되는 구조를 전제로 설계해, 외부 반출이 제한된 환경에서도 적용 가능하도록 한다.

# 3. AI 서비스 기획

## 서비스 시나리오

**(1) 생성 단계 시나리오**
```
═════════════════════ 생성 단계 ═════════════════════

     [회의 녹음]
          │
          │  (1) STT + 화자 분리
          ▼
     [발화록]    [U01] (14:00) 화자: 발화 내용
          │     [U02] (14:02) 화자: 발화 내용   ...
          │
          ▼
     (2) 회의록 작성자 Agent
          │      주제 / 일시 / 참석자 / 핵심 논의 / 결정사항  + [U##]
          │
          ▼  context
     (3) 액션아이템 및 예약 추출자 Agent
          │      담당자 | 할 일 | 기한  + [U##]
          │      일시 | 장소 | 목적     + [U##]
          │
          ▼  context
     (4) 인용 검증 편집자 Agent
          │      [U##] 실재성 대조 → 근거 없는 문장 삭제
          ▼
     [최종 회의록]
          │
          ├──▶ (5) audit_citations()
          │         정규식으로 [U##] 재대조 → 실재하지 않는 인용 검출
          │
          └──▶ (6) 예약 초안 ──▶ [사람 승인] ──▶ 예약 등록
                                      └ 미승인 시 등록되지 않음

   ※ (2)~(4) 각 Task에는 원본 발화록이 함께 주입된다.

```
1. 회의가 종료되면 녹음 파일이 STT와 화자 분리를 거쳐 발화록으로 변환된다. 각 발화에는 발화 ID, 시각, 화자가 부착된다.
2. 회의록 작성자 Agent가 주제, 일시, 참석자, 핵심 논의, 결정사항을 정리하고 각 문장에 근거 발화 ID를 붙인다.
3. 액션아이템 및 예약 추출자 Agent가 담당자와 기한이 포함된 액션아이템, 일시와 장소가 명시된 예약 초안을 생성한다.
4. 인용 검증 편집자 Agent가 모든 인용을 발화록과 대조하고 근거가 없는 문장을 제거한 뒤 최종 회의록을 출력한다.
5. 정규식 기반 감사 함수가 최종 산출물의 인용을 다시 대조해 실재하지 않는 인용을 검출한다.
6. 예약 초안은 확정되지 않은 상태로 전달되며, 사람이 승인해야 실제 등록이 이뤄진다.

**(2) 질의응답 단계 시나리오**

```
════════════════════ 질의응답 단계 ════════════════════

     (7) 사용자 질문 ──┐
                       ├──▶ (8) 회의 지식 응답자 Agent
        [발화록] ──────┘             │
                                     ├─ 근거 있음
                                     │    → 답변 + [U##]
                                     │       └─▶ audit_citations()
                                     │
                                     └─ (9) 근거 없음
                                          → "확인되지 않습니다"


─────────────────────────────────────────────────────
구현 범위  : (2)~(5), (7)~(9)
범위 밖    : (1) STT 및 화자 분리
                 → 결과물 형태를 재현한 샘플 발화록으로 대체
           (6) 사내 예약 시스템 등록
                 → 초안 생성까지만 구현
```
7. 사용자가 지난 회의에 대해 자연어로 질문한다.
8. 회의 지식 응답자 Agent가 발화록만을 근거로 답하고, 각 문장에 근거 발화 ID를 붙인다.
9. 발화록에서 근거를 찾을 수 없으면 추측하지 않고 확인되지 않는다고 답한다.

본 구현 범위는 2번부터 5번까지와 7번부터 9번까지다. STT와 화자 분리는 결과물 형태를 재현한 샘플 발화록으로 대체했다.

## 주요 기능

- **회의록 자동 생성**: 주제, 일시, 참석자, 핵심 논의, 결정사항을 구조화된 마크다운으로 생성
- **액션아이템 추출**: 담당자, 할 일, 기한을 표 형태로 정리
- **회의실 예약 초안 작성**: 일시, 장소, 목적을 명시해 승인만으로 처리 가능한 형태로 생성. 실제 등록은 사람의 승인을 거친다
- **이전 회의 내용 질의응답**: 발화록을 근거로 자연어 질문에 답하고 근거 인용을 함께 제시. 근거가 없으면 확인 불가로 응답
- **근거 인용 부착 및 검증**: 생성과 응답 양쪽에 인용을 부착하고, 생성 이후 코드로 실재 여부를 대조

## Biz 가치
**왜 필요한 서비스인가**

회의 음성의 외부 반출이 제한된 조직은 시중 제품을 도입할 수 없어 지금도 수기 정리에 의존한다. 이들에게는 가격이나 기능이 아니라 도입 가능 여부 자체가 문제다. 추론을 조직 내부에서 완결하는 구조는 기능 추가로 얻어지지 않고 처리 구조를 다시 설계해야 하므로, 이 지점이 기존 제품과의 차이가 된다.

**어떤 문제를 해결하는가**

- 회의록 수기 정리 시간을 줄인다. 회의 한 건당 20분에서 30분이 소요되던 작업을 대체한다.
- 후속 조치 누락을 방지한다. 액션아이템과 예약이 담당자, 기한, 일시, 장소가 명시된 형태로 정리되므로 별도 시스템 입력 과정에서 빠지지 않는다.
- 과거 논의를 찾는 비용을 줄인다. 녹음을 재생하는 대신 자연어 질문으로 확인하고, 답변에 붙은 근거 발화로 원문을 바로 대조할 수 있다.

**실제 활용 가능성**

CrewAI는 OpenAI 호환 엔드포인트로 모델을 호출하므로, `LLM(model=..., base_url=...)`의 두 값만 사내 서빙 엔드포인트로 교체하면 폐쇄망으로 전환된다. 본 구현은 검증 편의를 위해 외부 API로 실행했으나 코드 변경 없이 전환 가능한 구조를 유지했다. 다만 오픈웨이트 모델은 지시 준수도와 출력 형식 안정성이 상용 모델보다 낮은 경향이 있어, 인용 표기 규칙이 유지되는지는 실측 검증이 필요하다.

**기대효과**

기존 제품이 접근하지 못한 조직에서 회의록 자동화가 가능해진다. 또한 요약과 답변이 모두 원문 발화로 역추적되므로 회의 기록이 검증 가능한 형태로 축적되어, 감사 대응이나 인수인계 상황에서 결정의 근거를 제시할 수 있다.

# 4. 기술 구현

## 사용한 Transformer 기반 LLM

**모델**

- 데모: OpenAI `gpt-4o-mini` (litellm 형식 `openai/gpt-4o-mini`)
- 실배포 전제: 오픈웨이트 모델을 vLLM으로 사내 서빙하고 `base_url`과 `model`만 교체

**사용 목적**

비정형 회의 발화에서 결정사항과 액션아이템을 추출하고 요약을 생성한다. 결정은 "그러면 재고 조립체로 대체하겠습니다" 같은 문장으로 표현되고, 액션아이템은 "제가 등록해 두겠습니다"처럼 담당자가 생략된 채 말해진다. 무엇이 결정이고 누가 담당인지 판별하려면 문맥 이해가 필요하므로 규칙 기반으로는 처리할 수 없다.

**설정**

`temperature=0.0`으로 설정했다. 회의록은 표현의 다양성이 필요 없고 사실 보존과 인용 정확도가 최우선이므로, 확률 분포에서 최상위 토큰만 선택하도록 고정했다.

`verbose`는 기본값을 꺼두었다. CrewAI의 verbose는 발화록 전문과 프롬프트를 콘솔에 그대로 출력하는데, 보안 규제 조직에서는 이 로그 자체가 유출 경로가 되기 때문이다. 데모에서 사고 과정을 확인할 때만 환경변수로 켠다.

## Prompt Engineering

**Role Prompt**

네 Agent에 각각 `role`, `goal`, `backstory`를 부여했다. 역할이 명시되면 같은 입력에도 출력의 관점과 형식이 일관되게 유지된다.

| Agent | role | goal 요지 |
| --- | --- | --- |
| ① | 회의록 작성 전문가 | 발화록을 구조화하고 모든 요약 문장에 `[U##]` 부착 |
| ② | 후속 조치 관리자 | 액션아이템과 예약 필요사항을 승인 대기용 초안으로 정리 |
| ③ | 인용 검증 편집자 | 인용의 실재 여부를 대조하고 근거 없는 문장 삭제 |
| ④ | 회의 지식 응답자 | 발화록만을 근거로 답하고 근거 인용 부착 |

backstory에는 각 역할의 원칙을 적었다. ①에는 근거로 지목할 발화가 없는 문장은 쓰지 않는다는 원칙을, ④에는 근거가 없으면 확인되지 않는다고 답한다는 원칙을 명시했다.

**Few-shot**

회의록 작성 Task에 인용이 부착된 문장 예시 두 건을 제시했다. 인용 표기는 규칙으로 설명하는 것보다 예시로 보여주는 편이 형식 이탈이 적다.

**Output Format 지정**

각 Task의 `expected_output`에 출력 구조와 인용 규칙을 명시했다. 회의록은 정해진 순서의 마크다운으로, 액션아이템과 예약 초안은 각각 고정된 열을 가진 표로 지정했다. 형식을 앞 단계에서 고정해야 뒤의 검증과 코드 감사가 성립한다. 자유 서술이면 정규식 추출 자체가 불가능하다.

**단계별 Prompt 구성**

`Process.sequential`과 `context`로 요약, 추출, 검증을 각각 다른 Task로 분리했다. 하나의 프롬프트로 모두 처리하면 작업들이 서로의 품질에 영향을 주며, 특히 인용 부착은 요약 생성과 동시에 할 때 누락이 잦다. 각 Task의 description에는 요구사항을 번호로 나열해 수행 순서를 지정했다.

**금지 지시**

모든 Task에 발화록에 없는 내용을 생성하지 말 것과 근거 없는 문장을 출력하지 말 것을 명시했다. 질의응답 Agent에는 근거를 찾을 수 없을 때 출력할 문구까지 고정해, 모른다는 답을 회피하지 않도록 했다.

**프롬프트만으로 해결되지 않는 부분**

프롬프트 지시는 다음 토큰의 확률 분포를 유도할 뿐 출력을 강제하지 않는다. 존재하지 않는 발화 ID가 생성될 가능성이 남으므로, 생성 이후 `audit_citations()`로 정규식 대조를 수행하는 단계를 별도로 두었다. 확률적 생성의 결과를 결정론적 절차로 확인하는 구조다.

## AI Agent 구성

```
[생성 파이프라인 — 순차 Crew]

  회의 발화록 ([U##] 발화 ID, 시각, 화자 부착)
         |
         v
  ① 회의록 작성자 ── 주제/일시/참석자/핵심 논의/결정사항 + [U##]
         |  context
         v
  ② 액션아이템 및 예약 추출자 ── 담당자·기한 표, 예약 초안 표 + [U##]
         |  context
         v
  ③ 인용 검증 편집자 ── [U##] 대조, 근거 없는 문장 삭제
         v
     최종 회의록
         |
         v
     audit_citations() ── 인용 실재성 + 문장 근거율 코드 검증
                          (미통과 시 enforce 모드로 강제 제거)

[질의응답 — 온디맨드 단일 에이전트 Crew]

  질문 + 발화록 ──▶ ④ 회의 지식 응답자 ──▶ 답변 + [U##]
                                          또는 "확인되지 않습니다"
```

| Agent | 역할 |
| --- | --- |
| ① 회의록 작성자 | 발화록을 주제, 일시, 참석자, 핵심 논의, 결정사항으로 구조화하고 각 문장에 근거 발화 ID를 부착한다 |
| ② 액션아이템 및 예약 추출자 | 담당자와 기한이 포함된 액션아이템, 일시와 장소가 명시된 예약 초안을 승인 대기 형태로 정리한다 |
| ③ 인용 검증 편집자 | 앞 두 단계 산출물의 인용을 발화록과 대조하고 근거가 없거나 잘못된 문장을 제거한 뒤 병합한다 |
| ④ 회의 지식 응답자 | 발화록만을 근거로 질문에 답하고 근거 인용을 부착한다. 근거가 없으면 확인 불가로 응답한다 |

**구조 설계의 근거**

①에서 ③은 `practice_2`의 기획자, 작성자, 편집자 순차 구조를 계승하되 편집자를 인용 검증으로 특화했다. 생성과 검증을 같은 호출에서 처리하면 검증이 형식적으로만 수행되므로 독립 단계로 분리했다.

④는 생성 파이프라인과 별개로 동작한다. 생성은 회의 직후 한 번 실행되지만 질의응답은 임의 시점에 반복 호출되므로, 실행 주기가 달라 질문마다 단일 에이전트 Crew를 즉석에서 구성하는 방식으로 구현했다.

**다단계 구성의 비용**

Task마다 추론 파이프라인 전체가 처음부터 다시 수행된다. 토큰화, 임베딩, Self-Attention 연산이 매 호출마다 반복되고 발화록 전체가 다시 주입되므로 입력 토큰이 단계 수만큼 누적된다. 이 비용을 감수한 것은 검증 단계 분리가 산출물 신뢰도에 직접 영향을 주기 때문이다. 또한 외부 API 환경에서는 다단계 호출이 발화록의 반출 횟수를 늘리지만, 추론이 내부에서 완결되면 이 부담이 없다.

## 프로젝트 구현 내용

- CrewAI의 `Agent`, `Task`, `Crew(Process.sequential)`, `kickoff_async`로 생성 3에이전트 파이프라인과 질의응답 단일 에이전트를 구현했다.
- 발화 ID, 시각, 화자가 부착된 방산 진도점검 회의 발화록 12건을 입력으로 사용했다. STT와 화자 분리의 결과물 형태를 재현한 것이다.
- 각 Task에 원본 발화록을 함께 주입해 검증 단계에서 원문과 직접 대조할 수 있도록 했다.
- `audit_citations()`를 두 축으로 구성했다. 인용된 발화 ID가 발화록에 실재하는지 확인하는 축과, 근거가 붙어야 할 문장이 실제로 유효한 인용을 갖는지 확인하는 축이다. 후자는 불릿 항목과 표의 데이터 행만 대상으로 하며, 제목과 표 머리글, 구분선은 판정에서 제외한다.
- `enforce` 모드를 두어 검증에 실패하면 근거 없는 문장을 제거한 텍스트를 함께 반환한다. 인용 강제가 프롬프트 지시에 그치지 않고 코드 수준에서 보장된다.
- `ask_meeting(question)`으로 자연어 질문에 답한다. 실행 중인 이벤트 루프와 충돌하지 않도록 `kickoff_async` 기반 비동기 함수로 구현했다.
- `verbose`를 기본값 꺼짐으로 두어 발화록 원문이 콘솔과 로그로 출력되지 않도록 했다.

# 5. Transformer 이해

## Tokenization

입력 문장은 서브워드 단위 토큰으로 분리되어 모델에 들어간다. 본 프로젝트에서 문제가 되는 지점은 발화 ID 태그로, `[U01]`은 단일 토큰이 아니라 여러 서브워드로 분할되므로 형식이 불규칙하면 복원이 불안정해진다. 그래서 **발화 ID를 대문자 한 글자와 두 자리 숫자라는 고정 폭 형식으로 설계**했다. 또한 **한국어는 영어보다 토큰을 더 많이 소비**하므로, 한국어 회의를 다루는 본 서비스에서는 토큰 수가 곧 비용이자 처리 가능한 회의 길이의 한계가 된다.

---

## Embedding

임베딩은 토큰을 의미를 담은 벡터로 바꾸는 과정이며, 의미가 비슷한 토큰일수록 벡터 공간에서 가까이 놓인다. Transformer는 토큰 임베딩만으로 순서 정보를 갖지 못해 위치 정보를 별도로 더하는데, 회의록에서 발화 순서는 결정과 번복을 가르는 정보이므로 이 부분이 중요하다. **본 구현은 단일 회의가 대상이라 발화록 전체가 컨텍스트에 들어가므로 별도의 임베딩 검색을 쓰지 않았다.** 회의가 누적되면 전체를 매번 넣을 수 없으므로, **임베딩 검색은 컨텍스트 윈도우 제약을 우회하는 수단**으로 확장 단계에 배치했다.

---

## Self-Attention

Self-Attention은 각 토큰에서 Query, Key, Value를 만들어 Query와 Key의 내적으로 연관도를 구하고, 이를 가중치로 Value를 합해 표현을 갱신하는 연산이다. **순차적으로 은닉 상태를 전달하는 LSTM 계열과 달리 거리와 무관하게 직접 참조하므로 장기 의존성 문제를 구조적으로 해결한다.** 실행 결과에서도 세 발화만큼 떨어진 `[U02]`와 `[U05]`가 하나의 답변으로 결합되었고, "제가 등록해 두겠습니다"라는 발화의 담당자가 화자 정보로부터 복원되었다. 다만 **모든 토큰 쌍을 계산하므로 연산량이 입력 길이의 제곱에 비례**하며, 이것이 긴 회의를 한 번에 처리할 수 없는 원인이다.

---

## Next Token Prediction

모델은 다음 토큰의 확률 분포를 계산해 하나를 선택하고, 선택된 토큰이 다시 입력에 포함되는 자기회귀 방식으로 문장을 생성한다. `temperature=0.0`으로 설정한 것은 최상위 토큰만 선택하게 해 같은 입력에 같은 출력이 나오도록 하기 위한 것이다. 인용도 이 생성 과정에서 만들어지는 토큰이므로, **프롬프트 지시는 확률 분포를 유도할 뿐 출력을 강제하지 않아 존재하지 않는 발화 ID가 생성될 가능성이 원리적으로 남는다.** `audit_citations()`와 `enforce` 모드를 둔 이유가 여기에 있다.

---

## 프로젝트와 Transformer의 연관성

본 프로젝트의 설계 결정은 대부분 Transformer의 특성에서 파생되었다. 생성이 확률적 과정이라는 점 때문에 **인용 검증을 프롬프트가 아닌 코드로 이중화**했고, 태그가 여러 서브워드로 분할된다는 점 때문에 **발화 ID를 고정 폭 형식으로 설계**했다. 범위를 정하는 판단도 마찬가지여서, 단일 회의는 컨텍스트에 전부 들어가므로 **임베딩 검색을 확장 단계로 미뤘고**, 생략된 담당자와 기한을 복원하는 작업은 여러 발화를 건너뛴 참조가 필요하므로 **규칙 기반이 아닌 LLM에 맡기는 방식으로 설정을 하였다**.

# 6. 실행 결과

## 실행 화면

`python meeting_agent.py` 실행 시 다음 출력이 순서대로 표시된다.

- 최종 회의록 (회의록, 액션아이템 표, 예약 초안 표)
- 인용 검증 결과 (인용 실재성, 문장 근거율, 통과 여부)
- 질의응답 두 건과 각각의 인용 검증 결과

---

## 테스트 결과

**실행 환경**

- 모델: `openai/gpt-4o-mini`, `temperature=0.0`
- 입력: 발화 12건의 방산 진도점검 회의 발화록 (○○유도무기 구성품 3차 진도점검)

**생성 파이프라인**

세 Task가 순차 실행되어 핵심 논의 3건과 결정사항 6건을 담은 회의록, 담당자와 기한이 명시된 액션아이템 3행, 일시와 장소가 명시된 예약 초안 1행이 생성되었다. 모든 항목에 근거 발화 ID가 부착되었다.

```
인용      : 13개  (유효 13 / 환각 0)
청구 문장  : 13/13 근거 보유  (통과율 100%)
판정      : 통과 (모든 문장이 발화로 역추적됨)
```

환각 인용이 없었고 근거가 필요한 문장이 모두 유효한 인용을 가져, `enforce` 모드는 실행되지 않았다. 회의록의 일시 항목에는 발화록에 있는 시각만 기재되었고 날짜가 임의로 보충되지 않았다.

**질의응답**

근거가 있는 질문과 없는 질문 두 건으로 검증했다.

질문 1은 탐색기 조립체 납품 지연 사유와 대체 방안을 물었다. 답변은 형상 변경으로 2주 지연되었다는 내용을 `[U02]`로, 재고 조립체로 대체하고 정품은 2차 로트에 반영한다는 결정을 `[U05]`로 제시했다. 두 정보가 서로 다른 발화에 있음에도 하나의 답변으로 결합되었고, 인용 2건 모두 실재하는 ID로 확인되었다.

질문 2는 발화록에 없는 양산 물량을 물었다. 답변은 지어내지 않고 확인되지 않는다는 문구를 그대로 출력했다.

**확인된 동작**

- 세 Task가 오류 없이 순차 완료되며 `context`로 앞 단계 산출물이 전달된다.
- 담당자가 "제가"로만 표현된 발화에서 실제 담당자가 화자 정보로부터 복원되었다.
- 검증 함수가 제목과 표 머리글을 근거 필요 문장에서 제외해, 형식 요소를 오탐하지 않았다.
- 생성과 질의응답 양쪽 모두에서 인용이 실재하는 발화 ID로 역추적되었다.

# 7. 프로젝트 회고

## 잘된 점

**문제 정의를 요약 품질이 아닌 추론 위치로 잡았다**

시중 제품과 요약 성능을 겨루는 대신, 외부 반출이 제한되어 그 제품들을 쓸 수 없는 조직으로 범위를 좁혔다. 기능 경쟁이 아니라 도입 가능 여부라는 축이 생겨 기존 제품과 겹치지 않는 위치를 확보했다.

**인용 강제를 프롬프트가 아닌 코드로 보장했다**

프롬프트 지시만으로는 규칙 준수를 담보할 수 없으므로 생성 이후 정규식 대조를 붙이고, 검증에 실패하면 근거 없는 문장을 실제로 제거하는 강제 모드까지 두었다. 프롬프트로 해결되지 않는 부분을 코드로 보완한다는 역할 분담이 명확해졌다.

**검증을 두 축으로 나누고 형식 요소를 판정에서 제외했다**

인용된 ID가 실재하는지와 근거가 필요한 문장이 인용을 갖는지를 따로 계산했다. 제목과 표 머리글, 구분선은 대상에서 빼 형식 요소를 근거 없는 문장으로 오탐하지 않도록 했다.

**보안 논거를 코드에 반영했다**

`verbose`를 기본값으로 꺼 발화록 원문이 콘솔과 로그에 남지 않게 했다. CrewAI의 verbose는 프롬프트와 발화록 전문을 그대로 출력하므로, 이를 켜둔 채 배포하면 로그 자체가 유출 경로가 된다.

---

## 아쉬운 점

**인용 ID는 맞지만 내용이 다른 경우를 걸러내지 못한다**

검증이 ID의 실재 여부와 문장의 인용 보유 여부만 확인하므로, 실재하는 ID를 엉뚱한 문장에 붙이면 그대로 통과한다.

**질의응답에서는 문장 근거율 지표가 작동하지 않는다**

근거 필요 문장을 불릿과 표의 데이터 행으로 한정했는데, 답변은 산문이라 판정 대상이 0건으로 집계된다. 실행 결과에서도 인용은 2건 검출되었으나 청구 문장은 0/0으로 표시되었다.

**구현 범위의 한계**

STT와 화자 분리는 구현하지 않고 발화록을 입력으로 가정했다. 질의응답은 발화록 전체를 컨텍스트에 넣는 방식이며 임베딩 기반 검색은 넣지 않았고, 예약 초안도 생성까지만 수행하고 시스템 연동은 구현하지 않았다.

**검증 조건의 한계**

폐쇄망을 전제로 설계했으나 데모는 외부 API로 실행해 오픈웨이트 모델에서 인용 형식이 유지되는지 확인하지 못했다. 샘플 발화록 한 건으로만 검증했고, 인용 감사 정규식이 두 자리 발화 ID만 인식해 발화 100건이 넘는 회의에는 대응하지 못한다.

---

## 개선 방향

**검증 정확도**

- 인용된 발화와 문장의 의미 일치까지 대조하는 단계를 추가한다. 임베딩 유사도로 의미 거리를 계산해 기준값 이하이면 경고를 발생시키는 방식을 검토한다.
- 산문 형태의 답변도 문장 단위로 분리해 근거 보유 여부를 판정하도록 검증 기준을 확장한다.
- 정규식의 자릿수 제한을 해제해 발화 100건 이상의 회의에 대응한다.

**토큰 비용**

- 검증 단계에는 발화록 전체 대신 인용된 발화만 선별해 주입하는 방식으로 입력 토큰을 줄인다.
- 긴 회의는 발화록을 분할해 처리하고 결과를 병합하는 구조로 컨텍스트 윈도우 제약에 대응한다.

**폐쇄망 전환**

- 오픈웨이트 모델로 교체해 인용 표기 규칙이 유지되는지 실측한다.
- STT와 화자 분리를 포함한 전 구간을 사내에 구성해 통합 테스트를 수행한다.
- 화자 분리에 실패한 발화는 화자 없이 시각과 내용만 인용하도록 처리한다. 인용된 화자가 틀리면 근거 제시 기능이 오히려 신뢰를 떨어뜨리기 때문이다.

**기능 확장**

- 과거 회의록을 임베딩해 저장하고 관련 구간만 회수해 답변하는 구조로 질의응답을 확장한다. 이때 회의별 접근 권한으로 먼저 필터링한 뒤 검색해야 민감한 논의가 의도치 않게 노출되지 않는다.
- 예약 초안을 사내 예약 시스템 API와 연동해 승인 후 실제 등록까지 이어지도록 한다. 도구 호출이 도입되면 ReAct 구조를 적용할 수 있다.
